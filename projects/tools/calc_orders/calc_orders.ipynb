# MAIN ALGORITHM

## Ср.знач.продаж с учётом наличия

```python
# Расчёт средне недельного значения продаж за анализируемый период;
datamart.df['avg_sales'] = [
    round(
        np.nan_to_num(
            np.divide(sales, having) if having > 0 else sales,
            nan=0,
            posinf=0,
            neginf=0
        ),
        2
    )
    for sales, having in zip(
        datamart.df['ex_sale_cnt_sum_12_week'],
        datamart.df['hv_stock_12_week']
    )
]
```

```python
datamart.df
```

## ABC по бренду

```python
datamart.df = (
    datamart.df
    .merge(
        right=nomenclature,
        how='left',
        on=['brand_name_1c', 'sku_art_num']
    )
)
```

```python
datamart.df
```

```python
# Классификация ABC по продажам в количестве внутри брендов;
for index, row in tqdm(
        iterable=datamart.df[['brand_name_1c', 'parent_folder']].drop_duplicates().iterrows(),
        desc="Обработка брендов"
):
    brand, parent = row['brand_name_1c'], row['parent_folder']
    mask_brand_parent = datamart.df.query('brand_name_1c == @brand & parent_folder == @parent').index
    
    
    total_sum = datamart.df.loc[mask_brand_parent, 'ex_sale_cnt_sum_12_week'].sum()
    
    mask_sorted = (
        datamart.df.loc[mask_brand_parent, :].sort_values('ex_sale_cnt_sum_12_week', ascending=False).index
    )
    
    datamart.df.loc[mask_sorted, 'ratio'] = [
        np.nan_to_num(
            np.divide(sales, total_sum) if sales > 0 else np.divide(0.1, total_sum),
            nan=0.001,
            posinf=0.001,
            neginf=0.001,
        )
        for sales in datamart.df.loc[mask_sorted, "ex_sale_cnt_sum_12_week"]
    ]
    
    datamart.df.loc[mask_sorted, "cumsum"] = np.cumsum(
        datamart.df.loc[mask_sorted, "ratio"]
    )

    datamart.df.loc[mask_sorted, "abc"] = [
        "A" if value <= 0.75 else "B" if value <= 0.95 else "C"
        for value in datamart.df.loc[mask_sorted, "cumsum"]
    ]
```

```python
# Удалить временные столбцы;
datamart.df.drop(labels=['ratio', 'cumsum'], axis=1, inplace=True)
```

## Потребность

```python
# Расчёт потребности;
# Ср.знач продаж * срок запаса + срок доставки * ср.знач продаж + 2 (страховой коэфициент) * ср.знач продаж;

datamart.df.loc[:, 'requirement'] = [
    round(avg_sale * 12 + deliv * avg_sale + avg_sale * 2, 0)
    if abc == "A"
    else round(avg_sale * 8 + deliv * avg_sale + avg_sale * 2, 0)
    if abc == "B"
    else round(avg_sale * 4 + deliv * avg_sale + avg_sale * 2, 0)
    
    for avg_sale, abc, deliv in zip(
        datamart.df['avg_sales'],
        datamart.df['abc'],
        datamart.df['weeks_deliver']
    )
]
```

## Прогноз потребности;

```python
# Прогнозирование потребности на 52 недели вперед
datamart.df.loc[:, 'pred_requirement'] = [
    [req] * 52
    for req in datamart.df['requirement']
]
```

```python
datamart.df.sample(25)
```

## Предсказание остатка

```python
def predict_stock(total_stock: float, avg_sales: float) -> list:
    result: list = []
    for _ in range(52):
        result.append(max(round(total_stock - avg_sales), 0))
        total_stock -= avg_sales
    
    return result
```

```python
datamart.df['pred_stock'] = [
    predict_stock(
        total_stock=stock,
        avg_sales=sales
    )
    for stock, sales in zip(
        datamart.df['total_stock'],
        datamart.df['avg_sales']
    )
]
```

# Определение количества к заказу

```python
# Расчёт значения к заказу;
# 
datamart.df.loc[:, 'to_order'] = [
    [
        int(num) if num >= 0 else int(0)
        for num in np.subtract(np.array(req), np.array(stock))
    ]
    for req, stock in tqdm(
            zip(
            datamart.df['pred_requirement'],
            datamart.df['pred_stock']
        )
    )
]
```

```python
datamart.df.sample(25)
```

```python
datamart.df.loc[:, 'need_to_order'] = [
    int(order_array[int(weeks - 1)])
    for order_array, weeks, backorder in zip(
        datamart.df['to_order'],
        datamart.df['weeks_deliver'],
        datamart.df['cr_orders_backorder_cnt']
    )
]
```

```python
datamart.df.loc[:, 'need_to_order_without_backorder'] = [
    nto - bo
    for nto, bo in zip(
        datamart.df['need_to_order'],
        datamart.df['cr_orders_backorder_cnt']
    )
]
```
