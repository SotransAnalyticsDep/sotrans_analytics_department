```python
with datamart.engine.begin() as connection:
    match datamart.agg_cat_cols:
        case ('brand_name_1c', 'sku_art_num'):
            allias: str = 'ba'
        case ('wh_name', 'brand_name_1c', 'sku_art_num'):
            allias: str = 'wba'
    
    datamart.df.to_sql(
        name=f'{allias}_having',
        schema='metrics',
        con=connection,
        index=False,
        if_exists='replace'
    )
```
